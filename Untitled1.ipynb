```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

# Funo para carregar e processar os dados
def load_data(file_path):
    data = pd.read_csv(file_path, header=None, delim_whitespace=True)
    return data

# Funo para implementar o K-means manualmente
def load_data(file_path):
    data = pd.read_csv(file_path, header=None, sep='\s+')  # Correo: sep='\s+'
    return data

    for _ in range(max_iter):
        # Atribuir cada ponto ao cluster mais prximo
        for i in range(n_samples):
            distances = np.linalg.norm(data[i] - centroids, axis=1)  # Calculando a distncia
            labels[i] = np.argmin(distances)  # Atribui o cluster com menor distncia

        # Atualizar centrides
        new_centroids = np.array([data[labels == j].mean(axis=0) for j in range(k)])

        # Verificar convergncia (se os centrides no mudaram)
        if np.all(centroids == new_centroids):
            break
        centroids = new_centroids

    return centroids, labels

# Caminhos dos arquivos
train_path = "/content/X_train.txt"  # Substitua pelo caminho correto do arquivo X_train
test_path = "/content/X_test.txt"    # Substitua pelo caminho correto do arquivo X_test

# Carregar os dados de treino e teste
train_data = load_data(train_path)
test_data = load_data(test_path)

# Combinar os dados de treino e teste
data = pd.concat([train_data, test_data], axis=0).reset_index(drop=True)

# Normalizar os dados
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Reduzir a dimensionalidade com PCA para 2 componentes principais
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)

# Determinar o nmero ideal de clusters com o mtodo do cotovelo
inertia = []
silhouette_scores = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(data_pca)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(data_pca, kmeans.labels_))

# Plotar o mtodo do cotovelo e o Silhouette Score
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(range(2, 11), inertia, '-o')
plt.title('Mtodo do Cotovelo')
plt.xlabel('Nmero de Clusters (K)')
plt.ylabel('Inrcia')

plt.subplot(1, 2, 2)
plt.plot(range(2, 11), silhouette_scores, '-o', color='orange')
plt.title('Silhouette Score')
plt.xlabel('Nmero de Clusters (K)')
plt.ylabel('Pontuao')

plt.tight_layout()
plt.show()

# Executar K-means com o nmero ideal de clusters (ajustado para 4)
optimal_k = 4  # Escolha o K ideal com base nos grficos
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
labels = kmeans.fit_predict(data_pca)

# Visualizar os clusters em 2D
plt.figure(figsize=(8, 6))
for cluster in range(optimal_k):
    cluster_data = data_pca[labels == cluster]
    plt.scatter(cluster_data[:, 0], cluster_data[:, 1], label=f'Cluster {cluster}')

plt.title(f'Clusters Identificados (K={optimal_k})')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend()
plt.show()
```
